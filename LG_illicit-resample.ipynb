```python
import requests
from datetime import datetime
import time

import json
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
risk_dataset = pd.read_csv("dataset.csv")
```

```python
risk_dataset.describe().T 
```

```python

#StandardScaler

from sklearn.preprocessing import StandardScaler
numerical_col=['amount','fee','net_usage','multi_sign_fee','net_fee','energy_penalty_total','energy_usage','energy_fee','energy_usage_total','memoFee','origin_energy_usage','o_balance','o_transactions_in','o_transactions_out','o_totalTransactionCount','o_energyCost','o_netCost','o_freeNetUsed','o_age','t_balance','t_transactions_in','t_transactions_out','t_totalTransactionCount','t_energyCost','t_netCost','t_freeNetUsed','t_age']

scaler=StandardScaler()
scaled_numpy = scaler.fit_transform(risk_dataset[numerical_col])
scaled_df = pd.DataFrame(scaled_numpy, columns = numerical_col)

df_drop= risk_dataset.drop(columns = numerical_col)
risk_dataset=scaled_df.join(df_drop)  # StandardScaler
```

```python
df_risk_table_fix= risk_dataset.drop(columns=['hash', 'timestamp','ownerAddress','toAddress'])
df_risk_table_fix
```

```python
df_risk_table_fix_x=df_risk_table_fix.drop(columns=['illicit'])
df_risk_table_fix_y=df_risk_table_fix[['illicit']]
```

```python
df_risk_table_fix_x.describe().T 
```

```python
df_risk_table_fix_y
```

```python
df_risk_table_fix_y.value_counts()
```

```python
# re-sample by SMOTE
import imblearn
from imblearn.over_sampling import SMOTE

print("After over-sampling")

oversample = SMOTE()
x_resample, y_resample = oversample.fit_resample(df_risk_table_fix_x, df_risk_table_fix_y)   # over-sampling
# summarize the new class distribution

x_resample, y_resample = df_risk_table_fix_x, df_risk_table_fix_y # no over-sample


print(y_resample.value_counts() )
print(len(x_resample))
```

```python
x_resample.describe().T 
```

# LG


```python
## Base model

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score



acc_total=0
prec_total=0
f1_total=0
oob_total=0
auc_total=0
cvs_total=0
acc_best=0

prec_best=0
f1_best=0

auc_best=0
cvs_best=0
Acc_list = {}


set_estimators=100
iteration=20




for k in range( iteration):
   
    # Splitting the data into train and test
    X_train, X_test, y_train, y_test = train_test_split(x_resample, y_resample, train_size=0.7)
    
    classifier_lg = LogisticRegression(max_iter=1000,solver='liblinear')
    classifier_lg.fit(X_train, y_train.values.ravel())
    
    CV_score=cross_val_score(classifier_lg,X_train, y_train.values.ravel(),cv=5,scoring="accuracy")   
    cvs=np.average(CV_score)
    
    
    y_pred = classifier_lg.predict(X_test)
    print("\n")
    cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
    print("iteration :",k)
    print("(tn, fp, fn, tp) : " ,cnf_matrix.ravel())
    acc=metrics.accuracy_score(y_test, y_pred)
    print("Accuracy:",acc)
    prec=metrics.precision_score(y_test, y_pred,zero_division=0)
    print("Precision:",prec)
    f1=metrics.f1_score(y_test, y_pred,zero_division=0)
    print("F1 Score:",f1)

    print("CV_score : ", CV_score)
    print("CV_score average: ", cvs)    
    
    
    
        #define metrics
    y_pred_proba = classifier_lg.predict_proba(X_test)[::,1]
    fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
    auc = metrics.roc_auc_score(y_test, y_pred_proba)
    print("AUC :",auc)
    
    Acc_list[k] = acc
    
    acc_total=acc_total+acc
    prec_total=prec_total+prec
    f1_total=f1_total+f1
  
    auc_total=auc_total+auc
    cvs_total=cvs_total+cvs
    
    if cvs>cvs_best :
        cvs_best=cvs    
    if acc>acc_best :
        acc_best=acc
    if prec>prec_best :
        prec_best=prec
    if f1>f1_best :
        f1_best=f1

    if auc>auc_best :
        auc_best=auc

print("\n")
print("total iteration :", iteration)  

print("n estimators :", set_estimators)  

print("\n")
print("Accuracy average :",acc_total/iteration)
print("Precision average :",prec_total/iteration)
print("F1 average :",f1_total/iteration)

print("AUC average :",auc_total/iteration)
print("CV score average :",cvs_total/iteration)
print("\n")
print("Accuracy best :",acc_best)
print("Precision best :",prec_best)
print("F1 best :",f1_best)

print("CV score best :",cvs_best)
print("AUC best :",auc_best)

plt.figure()
plt.plot(list(Acc_list.keys()), list(Acc_list.values()))
plt.xlabel("iteration")
plt.ylabel("Acc_list")
#plt.show()
```

```python
from matplotlib import pyplot
importance = classifier_lg.coef_[0]
# summarize feature importance

imp_df = pd.DataFrame({
    "Varname": X_train.columns,
    "Imp": classifier_lg.coef_[0]
})

print(imp_df)
# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```

```python
## LGCV model

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegressionCV


acc_total=0
prec_total=0
f1_total=0
oob_total=0
auc_total=0

acc_best=0

prec_best=0
f1_best=0

auc_best=0

Acc_list = {}


set_estimators=100
iteration=20




for k in range( iteration):
   
    # Splitting the data into train and test
    X_train, X_test, y_train, y_test = train_test_split(x_resample, y_resample, train_size=0.7)
    
    classifier_lgcv = LogisticRegressionCV(cv=5, scoring="accuracy", n_jobs=-1,max_iter=1000)
    classifier_lgcv.fit(X_train, y_train.values.ravel())
    
    
    
    y_pred = classifier_lgcv.predict(X_test)
    print("\n")
    cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
    print("iteration :",k)
    print("(tn, fp, fn, tp) : " ,cnf_matrix.ravel())
    acc=metrics.accuracy_score(y_test, y_pred)
    print("Accuracy:",acc)
    prec=metrics.precision_score(y_test, y_pred)
    print("Precision:",prec)
    f1=metrics.f1_score(y_test, y_pred)
    print("F1 Score:",f1)

 
    
    
    
        #define metrics
    y_pred_proba = classifier_lgcv.predict_proba(X_test)[::,1]
    fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
    auc = metrics.roc_auc_score(y_test, y_pred_proba)
    print("AUC :",auc)
    
    Acc_list[k] = acc
    
    acc_total=acc_total+acc
    prec_total=prec_total+prec
    f1_total=f1_total+f1
  
    auc_total=auc_total+auc
  
    if acc>acc_best :
        acc_best=acc
    if prec>prec_best :
        prec_best=prec
    if f1>f1_best :
        f1_best=f1

    if auc>auc_best :
        auc_best=auc

print("\n")
print("total iteration :", iteration)  

print("n estimators :", set_estimators)  

print("\n")
print("Accuracy average :",acc_total/iteration)
print("Precision average :",prec_total/iteration)
print("F1 average :",f1_total/iteration)

print("AUC average :",auc_total/iteration)

print("\n")
print("Accuracy best :",acc_best)
print("Precision best :",prec_best)
print("F1 best :",f1_best)
print("AUC best :",auc_best)

plt.figure()
plt.plot(list(Acc_list.keys()), list(Acc_list.values()))
plt.xlabel("iteration")
plt.ylabel("Acc_list")
#plt.show()
```

```python
from matplotlib import pyplot
importance = classifier_lgcv.coef_[0]
# summarize feature importance

imp_df = pd.DataFrame({
    "Varname": X_train.columns,
    "Imp": classifier_lgcv.coef_[0]
})

print(imp_df)
# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```

# GridSearchCV


```python
# GridSearchCV


X_train, X_test, y_train, y_test = train_test_split(x_resample, y_resample, train_size=0.7)


LG = LogisticRegression(random_state=42, n_jobs=-1)

params = {
    'solver': ['newton-cg', 'lbfgs', 'liblinear'],
    'penalty': ['l1', 'l2'],
    'C': [1, 10, 100]
}
from sklearn.model_selection import GridSearchCV
# Instantiate the grid search model
grid_search = GridSearchCV(estimator=LG,
                           param_grid=params,
                           cv = 5,
                           n_jobs=-1, verbose=1, scoring="accuracy")

grid_search.fit(X_train, y_train.values.ravel())

y_pred = grid_search.predict(X_test)
print ('CV: ', grid_search, '\n')

lg_best = grid_search.best_estimator_
print ('Best Estimator: ', lg_best, '\n')

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
print("(tn, fp, fn, tp) : " ,cnf_matrix.ravel())
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("F1 Score:",metrics.f1_score(y_test, y_pred))
print("Best Score:",grid_search.best_score_)

#define metrics
y_pred_proba = grid_search.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
print("AUC :",auc)

print ( '\n')
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))

#create ROC curve
plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.legend(loc=4)
plt.show()
```
